```python
def generate_format(text):
    # """

    # Please generate meaningful and structured text from the given unstructured text using OpenAI API, removing directories, references, and meaningless or useless sentences:
    # print(f"prompt = {prompt}")
    """
    Please generate meaningful and structured text from the given unstructured text using OpenAI API, no changing the text, remaining all meaningful contents, and just removing directories, references, and meaningless sentences.


    """
    # Prompt for the model
    prompt = f"""
    Please generate meaningful and structured text from the given unstructured text using OpenAI API, no changing the text, remaining all meaningful contents, and just removing directories, references, and meaningless sentences:
    Text: "{text}"
    """
    try:
        # Call OpenAI API
        response = client.chat.completions.create(
            # model="gpt-3.5-turbo-instruct",
            model="gpt-4o", # gpt-4
            messages=[
                {"role": "system", "content": "You are a helpful assistant and text generator."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.7,
            max_tokens=4096
        )

        # Extract the response text
        # output = response['choices'][0]['message']['content']
#         output = response.choices[0].text
#         choices = response.choices
# chat_completion = choices[0]
# content = chat_completion.message.content
        output = response.choices[0].message.content
        return output

    except Exception as e:
        print(f"Error: {e}")
        return None
```

```python
# Split the text into smaller chunks

from langchain.text_splitter import RecursiveCharacterTextSplitter

def split_text_for_llm(text, chunk_size = 20480, chunk_overlap=2):
    """Splits a long text into smaller chunks suitable for LLM processing."""

    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
        length_function=len,
        separators=["\n\n", "\n", " ", ""]
    )

    chunks = text_splitter.split_text(text)
    return chunks
```

```python
chunks = split_text_for_llm(context)
print(chunks)
print(len(chunks))
```

```python
!pip install language_tool_python
import language_tool_python
tool = language_tool_python.LanguageTool('en-US')
```

```python
# Filter chunks based on grammar and spelling errors. Only keep chunks with fewer errors than the average.

def filter_chunks(chunks):
    chunk_score = {}
    scores = []
    i = 1
    for chunk in chunks:
        score = len(tool.check(chunk))
        scores.append(score)
        chunk_score[chunk] = score
        print(f"i = {i}, score = {score}")
        i += 1

    filtered_chunks = []
    score_avg = sum(scores) / len(scores)
    for chunk, score in chunk_score.items():
      if score < score_avg:
        filtered_chunks.append(chunk)
    return filtered_chunks

filtered_chunks = filter_chunks(chunks)
print(len(filtered_chunks))
```

```python
# Generate Q&A pairs
# qa_pairs = generate_qa_pairs(text)
# print(qa_pairs)
texts = ""
count=0
for chunk in filtered_chunks:
    try:
        txt = generate_format(str(chunk))
        print(f"txt = {txt}")
        count += 1
        print(f"count = {count}")
    except:
        print(f"No text from {chunk}")
    else:
        texts += str(txt)
    # break
print(texts)
```

```python
def saveTextToFile(data, outputFile):
    # print(data)
    with open(outputFile, "w") as outfile:
        outfile.write(data)
    outfile.close()

saveTextToFile(texts, '/content/drive/MyDrive/EngineeringReference1_5.txt')
```
